Ici une fonction polynome du troisieme degre: f(x)=x^3+2x+1.
    On veut, pour une valeur initiale de x choisie, un pas choisi, un nombre de valeurs de x choisi, l'affichage de toutes les images f(x).

```python
L=[]                                          # L est la liste dans laquelle figureront les valeurs de f(x)
debut=float(input('valeur initiale'))     # initialisation de x
pas=float(input('pas'))                   # on fixe un pas pour les valeurs prises par x
n=int(input('nombre de valeurs'))         # on fixe le nombre n d'images voulues
for k in range(n):                            # car on veut n images
    x=debut+k*pas                             # le pas est ajoute a x
    f=x**3+2*x+1                              # l'image f(x) est alors calculee
    L.append(f)                               # la valeur de f(x) est mise dans la liste
print(L     )                                  # la liste L est affichee
```

# Exemple

Soit une fonction polynome du troisieme degre: f(x)=x^3+2x+1.

Le tableau de variation montre que l'equation f(x)=0 admet une solution alpha unique et que -1<alpha<0.

On veut trouver un encadrement de alpha d'amplitude choisie

```python
def fonction(x):
    f=x**3+2*x+1
    return f

# methode par balayage
x=-1
amp=0.01
print("on cherche les solutions de f(x)=0 par balayage à partir de "+str(x)+" avec un pas de "+str(amp))
#amp=float(input('amplitude voulue'))
pas=amp
while fonction(x)<0: 
    x=x+pas
print(x-pas,'<alpha<',x)


# methode par dichotomie
a=-1.0
b=0
amp=0.01

print("on cherche les solutions de f(x)=0 par dichotomie entre "+str(a)+" et "+str(b))
#amp=float(input('amplitude voulue'))
while (b-a)>amp:
    m=(a+b)/2
    if fonction(m)<0: a=m
    else: b=m
print(a,'<alpha<',b)
```

# Comparatif dichotomie et newton
